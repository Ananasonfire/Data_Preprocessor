```python
from sklearn.datasets import fetch_openml
import pandas as pd
import sys
sys.path.append('..')  
from src.data_preprocessor import DataPreprocessor

print("Загрузка Titanic...")
titanic = fetch_openml('titanic', version=2, as_frame=True, parser='auto')['frame']
print(f"Shape: {titanic.shape}")
print("Пропуски:\n", titanic.isnull().mean().sort_values(ascending=False))
```

```python
print("Типы колонок:")
print(titanic.dtypes.value_counts())
print("\nПервые 5 строк:")
titanic.head()
```

```python
prep = DataPreprocessor(titanic)
df_clean = prep.fit_transform(threshold=0.5, method='std')
print(f" После обработки: {df_clean.shape}")
print(f"Удалено колонок: {prep.dropped_cols_}")
print("Лог:", prep.get_log())
```

```python
print("Первые 5 строк после OHE + нормализации:")
df_clean.head()
```

```python
test_df = titanic.sample(100, random_state=42)
df_test = prep.transform(test_df)
print(f"Transform работает: {df_test.shape == df_clean.shape}")
print("Колонки совпадают:", len(set(df_test.columns) & set(df_clean.columns)) == len(df_clean.columns))
```

```python
print("ФИНАЛЬНЫЕ РЕЗУЛЬТАТЫ:")
print(f"До:  {titanic.shape}")
print(f"После:{df_clean.shape}")
print(f"Новых OHE колонок: {len(prep.ohe_cols_)}")
print(f"Удалено: {len(prep.dropped_cols_)}")

```
